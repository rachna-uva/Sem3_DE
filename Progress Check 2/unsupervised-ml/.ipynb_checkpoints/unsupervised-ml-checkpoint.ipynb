## Step 3 : Feature Engineering
## 3.1 : Feature Creation

### Next, we deal with the features. We adopt a simple and classic approach, i.e. we calculate the following 3 factors:
- Recency: Number of days since last purchase
- Frequency: Number of tracsactions
- Monetary: Total amount of transactions (revenue contributed)

You can run the following code to compute the attribute "Recency".

```python
# New Attribute : Recency

# Compute last transaction date to get the recency of customers
max_date = max(retail['InvoiceDate'])
retail['Recency'] = max_date - retail['InvoiceDate']

# Compute recency by grouping the same Customers together
rfm_p = retail.groupby('CustomerID')['Recency'].min()
rfm_p = rfm_p.reset_index()

# Extract number of days only
rfm_p['Recency'] = rfm_p['Recency'].dt.days
rfm_p.head()
```

### Task 3
Compute the other two attributes: "Monetary" and "Frequency".

Quantity times the UnitPrice is the amount of one order and you can sum up all orders for the same CustomerID to get the Monetary, save this in a `df` called `rfm_m` in column 'Amount'.

You can count how many InvoiceNo for one CustomerID to get the Frequency, save this in a `df` called `rfm_f` in column 'Frequency'. 

```python
# YOUR CODE HERE
retail['Amount'] = retail['Quantity']*retail['UnitPrice']
rfm_m = retail.groupby('CustomerID')['Amount'].sum().reset_index()                      
#raise NotImplementedError()

# YOUR CODE HERE
rfm_f = retail.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
rfm_f.rename(columns = {'InvoiceNo':'Frequency'}, inplace = True)
#raise NotImplementedError()
```

```python
# Check if columns exist in rfm_m and rfm_f
assert 'CustomerID' in rfm_m.columns, "Expected column name: CustomerID"
assert 'Amount' in rfm_m.columns, "Expected column name: Amount"
assert 'CustomerID' in rfm_f.columns, "Expected column name: CustomerID"
assert 'Frequency' in rfm_f.columns, "Expected column name: Frequency"
```

### Task 4
Merge all three attributes to form a `dataframe` called "rfm".

```python
# YOUR CODE HERE
rfm = pd.merge(rfm_p, rfm_m, on = 'CustomerID', how = 'inner')
rfm = pd.merge(rfm, rfm_f, on = 'CustomerID', how = 'inner')
print(len(rfm))
#raise NotImplementedError()
```

## Step 4 : Building the Model
Now it is time to cluster the customers into groups. We use k-mean and hierachical clustering to achieve this task. 
We will be calculating the silhoute score for various settings and saving this in a dictionary.

### Task 8
Perform kmean clustering on the data with the number of clusters set to be 3.

And compute the Silhouette Scores.

```python
# Initialize an empty dictionary
silhouette_scores = {}
```

```python
# YOUR CODE HERE
# KMeans model
def kmeans_clustering(rfm_df_scaled, clusters):
    kmeans = KMeans(n_clusters = clusters, random_state = 42, n_init = 'auto')
    kmeans.fit(rfm_df_scaled)
    cluster_labels = kmeans.fit_predict(rfm_df_scaled)
    silhouette_avg = silhouette_score(rfm_df_scaled, cluster_labels)
    dict_key = 'Kmean' + str(clusters)
    silhouette_scores[dict_key]= silhouette_avg

    # display
    """
    print(f'Silhouette scores: \n{silhouette_scores}')
    print(f'Cluster labels: \n{cluster_labels}')
    print(f'Number of clusters: {len(cluster_labels)}')
    """
    return silhouette_scores, cluster_labels

silhouette_scores, cluster_labels = kmeans_clustering(rfm_df_scaled, 3)
rfm_df_scaled['KmeansClusterLabels3'] = cluster_labels
print(silhouette_scores)
#raise NotImplementedError()
```

### Task 9
Perform hierachical clustering on the data, the algorithm can be chosen as agglomerative clustering, number of clusters is 3, linkage is "ward".

Compute the Silhouette Scores.

```python
# YOUR CODE HERE
def hierarchical_clustering(rfm_df_scaled, clusters):
    agg_cluster = AgglomerativeClustering(n_clusters = clusters, linkage='ward')
    agg_cluster_labels = agg_cluster.fit_predict(rfm_df_scaled)
    silhouette_avg = silhouette_score(rfm_df_scaled, agg_cluster_labels)
    dict_key = 'Ward' + str(clusters)
    silhouette_scores[dict_key] = silhouette_avg
    # display
    """
    print(f'Silhouette scores: \n{silhouette_scores}')
    print(f'Aggregated cluster labels: \n{agg_cluster_labels}')
    print(f'Number of aggregated clusters: {len(agg_cluster_labels)}')
    """
    return silhouette_scores, agg_cluster_labels

silhouette_scores, cluster_labels = hierarchical_clustering(rfm_df_scaled, 3)
rfm_df_scaled['HierarchicalClusterLabels3'] = cluster_labels
print(silhouette_scores)
#raise NotImplementedError()
```

### Additional Task
You can use the scatter plot (Amount vs. Frequency) to show the differences between these two methods.

By observing the plot we know that the kmean shows clearer boundaries. 

According to their mechanism, do you know why?

```python
# YOUR CODE HERE
# Scatter plot for K-means clustering
def kmeans_plotting(rfm_df_scaled, clusters):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(rfm_df_scaled['Amount'], rfm_df_scaled['Frequency'], c = rfm_df_scaled[f'KmeansClusterLabels{clusters}'], cmap = 'viridis', edgecolors = 'k', alpha=0.7)
    plt.title(f'K-Means Clustering (n={clusters})')
    plt.xlabel('Amount')
    plt.ylabel('Frequency')

    plt.show()

kmeans_plotting(rfm_df_scaled, 3)
kmeans_plotting(rfm_df_scaled, 4)
#raise NotImplementedError()
```

```python
# YOUR CODE HERE
# Scatter plot for Agglomerative clustering
def hierarchical_plotting(rfm_df_scaled, clusters):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 2)
    plt.scatter(rfm_df_scaled['Amount'], rfm_df_scaled['Frequency'], c = rfm_df_scaled[f'HierarchicalClusterLabels{clusters}'], cmap = 'viridis', edgecolors='k', alpha=0.7)
    plt.title(f'Agglomerative Clustering (n={clusters})')
    plt.xlabel('Amount')
    plt.ylabel('Frequency')

    plt.show()
    
hierarchical_plotting(rfm_df_scaled, 3)
hierarchical_plotting(rfm_df_scaled, 4)
#raise NotImplementedError()
```

ANSWER: Agglomerative clustering, by default, may assign datapoints to multiple clusters, while K-Means clustering assigns each datapoint to only one cluster. Hence, the latter results in more clear and distinct boundaries between clusters in the scatter plot.
